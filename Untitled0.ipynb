```python
import os

# Paths
images_path = "/content/drive/MyDrive/vegetation segmentation.v2i.yolov8/train/images"
labels_path = "/content/drive/MyDrive/vegetation segmentation.v2i.yolov8/train/labels"

image_files = [f.split('.')[0] for f in os.listdir(images_path) if f.endswith(('.jpg', '.png'))]
label_files = [f.split('.')[0] for f in os.listdir(labels_path) if f.endswith('.txt')]

missing_labels = set(image_files) - set(label_files)
missing_images = set(label_files) - set(image_files)

print("✅ Total images:", len(image_files))
print("✅ Total labels:", len(label_files))
print("⚠️ Missing labels for:", missing_labels)
print("⚠️ Missing images for:", missing_images)
```

```python
def check_labels(labels_path):
    bad_files = []
    for file in os.listdir(labels_path):
        if file.endswith(".txt"):
            with open(os.path.join(labels_path, file), "r") as f:
                for line in f.readlines():
                    parts = line.strip().split()
                    if len(parts) != 5:
                        bad_files.append(file)
                    else:
                        try:
                            vals = list(map(float, parts[1:]))
                            if not all(0 <= v <= 1 for v in vals):
                                bad_files.append(file)
                        except:
                            bad_files.append(file)
    return bad_files

bad = check_labels(labels_path)
print("⚠️ Bad label files:", bad if bad else "None, all good ✅")
```

```python
import os


def count_images(folder_path):
    total = 0
    for root, dirs, files in os.walk(folder_path):
        total += len([f for f in files if f.lower().endswith(('.jpg', '.jpeg', '.png'))])
    return total

train_path = "/content/drive/MyDrive/vegetation segmentation.v2i.yolov8/train"
test_path  = "/content/drive/MyDrive/vegetation segmentation.v2i.yolov8/test"
valid_path = "/content/drive/MyDrive/vegetation segmentation.v2i.yolov8/valid"

print("Train images:", count_images(train_path))
print("Test images:", count_images(test_path))
print("Validation images:", count_images(valid_path))
```

```python

!pip install -U ultralytics albumentations --quiet
import os, glob, random, shutil
from IPython.display import display, Image

data_yaml = "/content/drive/MyDrive/vegetation segmentation.v2i.yolov8/data.yaml"
val_dir   = "/content/drive/MyDrive/vegetation segmentation.v2i.yolov8/valid/images"
test_dir  = "/content/drive/MyDrive/vegetation segmentation.v2i.yolov8/test/images"
backup_dir = "/content/drive/MyDrive/vegetation_segmentation_backups"
os.makedirs(backup_dir, exist_ok=True)


run_dirs = sorted([d for d in glob.glob('runs/*/*') if os.path.isdir(d)], key=os.path.getmtime)
if len(run_dirs) == 0:
    raise SystemExit("No training run found under runs/*/* — train first or point to a run folder.")
latest_run = run_dirs[-1]
print("Latest run folder:", latest_run)


weights_path = os.path.join(latest_run, "weights", "best.pt")
if not os.path.exists(weights_path):
    pts = sorted(glob.glob(os.path.join(latest_run, "*.pt")) + glob.glob(os.path.join(latest_run, "weights/*.pt")), key=os.path.getmtime)
    if len(pts) == 0:
        raise SystemExit("No .pt weights found in the run folder.")
    weights_path = pts[-1]
print("Using weights:", weights_path)


print("\n--- Training plots (showing up to 3) ---")
pngs = sorted(glob.glob(os.path.join(latest_run, "*.png")) + glob.glob(os.path.join(latest_run, "*/*.png")))
if len(pngs) == 0:
    print("No plot PNGs found in", latest_run)
else:
    for p in pngs[:3]:
        print("  ", p)
        display(Image(filename=p))

```

```python
print("\n--- Running inference on a few validation images ---")
model = YOLO(weights_path)

val_images = sorted(glob.glob(os.path.join(val_dir, "*.*")))
if len(val_images) == 0:
    print("No validation images found at:", val_dir)
else:
    sample_count = min(5, len(val_images))
    sample_images = random.sample(val_images, sample_count)
    print(f"Selected {sample_count} images for sample inference.")

    infer_save_dir = os.path.join(latest_run, "inference_val_samples")
    os.makedirs(infer_save_dir, exist_ok=True)


    for im in sample_images:
        print("Inferencing:", im)
        model.predict(source=im, save=True, save_dir=infer_save_dir, conf=0.25, imgsz=640)

    saved = sorted(glob.glob(os.path.join(infer_save_dir, "*.*")))
    print("Saved inference outputs (showing up to 5):", saved[:5])
    for s in saved[:5]:
        display(Image(filename=s))
```
